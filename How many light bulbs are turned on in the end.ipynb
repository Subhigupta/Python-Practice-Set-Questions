```python
#importing libraries
import pandas as pd
from pandas import DataFrame

# Finding the number (i.e nth bulb) is divisible by which numbers
def generate_factors(n,data):
    count=0;
    divisors=[]
    status=""
    for i in range(1,n+1):
        if n%i==0:
            count=count+1 #Keeping a count of factors
            divisors.append(i)
    if count%2==0:
        status="Off" #If total factors are even then the bulb is Off
    else:
        status="On" #If total factors are even then the bulb is On
    data=data.append({'nth Bulb':n,'Divisors of nth bulb':divisors,'Count of divisors of nth bulb':count,'On/Off status':status},ignore_index=True)
    return data

data=pd.DataFrame(columns=["nth Bulb","Divisors of nth bulb","Count of divisors of nth bulb","On/Off status"])
for i in range (1,101):
    data=generate_factors(i,data) # Calling the function and passing the nth bulb and dataframe
```
